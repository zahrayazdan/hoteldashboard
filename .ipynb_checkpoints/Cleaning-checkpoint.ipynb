```python
# Selecting only numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Summary statistics for numerical columns
numerical_summary = numeric_df.describe().transpose()

# Summary statistics for categorical columns
categorical_summary = df.describe(include=['object']).transpose()

# Missing values count
missing_values = df.isnull().sum().to_frame(name='Missing Values')
missing_values['% Missing'] = (missing_values['Missing Values'] / len(df)) * 100

# Data types of columns
data_types = df.dtypes.to_frame(name='Data Type')

# Unique values count for categorical columns
unique_values = df.select_dtypes(include=['object']).nunique().to_frame(name='Unique Values')

# Checking for duplicate rows
duplicate_rows = df.duplicated().sum()

# Outlier detection using Interquartile Range (IQR) - only for numerical columns
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).sum().to_frame(name='Potential Outliers')

# Displaying results
print("\n=== Numerical Summary ===\n", numerical_summary)
print("\n=== Categorical Summary ===\n", categorical_summary)
print("\n=== Missing Values ===\n", missing_values)
print("\n=== Data Types ===\n", data_types)
print("\n=== Unique Values in Categorical Columns ===\n", unique_values)
print(f"\n=== Number of Duplicate Rows: {duplicate_rows} ===")
print("\n=== Potential Outliers Per Column ===\n", outliers)
```

```python
# Handling Missing Values
# Fill missing 'country' values with the most common country
df['country'].fillna(df['country'].mode()[0], inplace=True)

# Filling missing 'agent' values with 0 (assuming it's an ID and 0 means 'Not Provided')
df['agent'].fillna(0, inplace=True)

# 'Company' column has 94.3% missing values; dropping it:
df.drop(columns=['company'], inplace=True)

# Fixing Data Types
# Converting 'agent' to integer (since it's an ID)
df['agent'] = df['agent'].astype(int)

# Converting 'children' to integer
df['children'] = pd.to_numeric(df['children'], errors='coerce').fillna(0).astype(int)

# Converting 'reservation_status_date' to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Handling Outliers
# Cap 'lead_time' at the 95th percentile to remove extreme long bookings
df['lead_time'] = df['lead_time'].clip(upper=df['lead_time'].quantile(0.95))

# Removing unrealistic values for 'adults' (e.g., bookings with more than 6 adults)
df = df[df['adults'] <= 6]

# Remove negative 'adr' (average daily rate) values
df = df[df['adr'] > 0]

# Standardize Categorical Values
# Standardize 'meal' column by replacing 'Undefined' with 'SC' (Self Catering)
df['meal'] = df['meal'].replace({'Undefined': 'SC'})

# Display final dataset info after cleaning
print(df.info())

# Save cleaned dataset (optional)
df.to_csv("cleaned_hotel_booking_dataset.csv", index=False)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\zahra\MBAS902 assignment1\cleaned_hotel_booking_dataset.csv")

### 1. Create Stay Duration Column ###
df["total_stay"] = df["stays_in_week_nights"] + df["stays_in_weekend_nights"]

### 2. Calculate Total Revenue ###
df["total_revenue"] = df["adr"] * df["total_stay"]

### 3. Extract Booking Month and Define Season ###
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"])
df["booking_month"] = df["reservation_status_date"].dt.month_name()

# Define seasons based on months
season_mapping = {
    "December": "Winter", "January": "Winter", "February": "Winter",
    "March": "Spring", "April": "Spring", "May": "Spring",
    "June": "Summer", "July": "Summer", "August": "Summer",
    "September": "Fall", "October": "Fall", "November": "Fall"
}
df["booking_season"] = df["booking_month"].map(season_mapping)

### 4. Create Cancellation Flag ###
df["cancellation_flag"] = df["is_canceled"].apply(lambda x: 1 if x == 1 else 0)

### 5. Categorize Lead Time ###
df["lead_time_category"] = pd.cut(df["lead_time"],
                                  bins=[0, 30, 90, df["lead_time"].max()],
                                  labels=["Short (<30 days)", "Medium (30-90 days)", "Long (>90 days)"])

### 6. Group Customer Type ###
customer_mapping = {
    "Transient": "Short-term traveler",
    "Contract": "Corporate booking",
    "Group": "Event traveler",
    "Transient-Party": "Short-term traveler"
}
df["customer_segment"] = df["customer_type"].map(customer_mapping)

# Save the modified dataset
df.to_csv("final_hotel_booking.csv", index=False)

# Display confirmation message
print("New columns added and dataset saved as 'final_hotel_booking.csv'")
```
